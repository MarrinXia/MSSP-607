```python
from google.colab import drive
drive.mount('/content/drive')

# To match Drive folder
path = '/content/drive/MyDrive/Week5 Examples'
import pandas as pd
import glob
import os
# Find all CSVs inside the folder
files = glob.glob(os.path.join(path, '*.csv'))

dfs = []
for f in files:
    ticker = os.path.basename(f).split('.')[0].upper()   # e.g. 'AAPL'
    df = pd.read_csv(f)
    df['ticker'] = ticker
    dfs.append(df)

# Combine all into one dataframe
faanf = pd.concat(dfs, ignore_index=True)
faanf.head()
faanf.to_csv(os.path.join(path, 'faanf.csv'), index=False)

faanf['date'] = pd.to_datetime(faanf['date'])
faanf['volume'] = faanf['volume'].astype(int)

faanf = faanf.sort_values(by=['date', 'ticker'])
faanf.head()

top7 = faanf.nlargest(7, 'volume')
print(top7)

melted = pd.melt(faanf, id_vars=['date', 'ticker'])
melted.head()


```
